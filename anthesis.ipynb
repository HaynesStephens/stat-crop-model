```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
daydir = '/Users/haynes/Downloads/pdssat/'
```

```python
ds = xr.load_dataset(daydir+'pdssat_agmerra_fullharm_anth-day_mai_global_annual_1980_2010_C360_T0_W0_N200_A0.nc4', 
                     decode_times=False).sel(lon=slice(-104,-80)).sel(lat=slice(49,36))
ds['time'] = np.arange(1980,2011)
anth = ds['anth-day_mai'].mean(dim='time')
print(anth.min(), anth.max())
# anth.plot(figsize=(12,8))
```

```python
ds = xr.load_dataset(daydir+'pdssat_agmerra_fullharm_anth-day_mai_global_annual_1980_2010_C360_T6_W0_N200_A0.nc4', 
                     decode_times=False).sel(lon=slice(-104,-80)).sel(lat=slice(49,36))
ds['time'] = np.arange(1980,2011)
anth2 = ds['anth-day_mai'].mean(dim='time')
print(anth2.min(), anth2.max())
# anth.plot(figsize=(12,8))
```
